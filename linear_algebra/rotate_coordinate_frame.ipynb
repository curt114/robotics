```python
# Imports
import sympy as sp
from sympy.physics.vector import *
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
sp.init_printing()

'''
    1. Rotate a point (vector)
    2. Counterclockwise by angle theta.
''' 

# World Reference Frame and rotation matrix coordinate frame.
W = ReferenceFrame('W')
R = ReferenceFrame('R')

# Create the symbolic symbol for theta.
theta = sp.symbols('theta')

# Orientate Reference frame R from the base frame W.  What type of rotation - axis
R.orient(W, 'Axis', [theta, W.z])

# Reference the new vectors in the reference frame instead of the world frame.
R.dcm(W)

def angle(rx, ry, rotation):
    R.orient(W, 'Axis', [rotation, W.z])
    return R.dcm(W) * sp.Matrix([[rx], [ry], [0]])
```

```python
angle(5, 4, 0.524)
```
